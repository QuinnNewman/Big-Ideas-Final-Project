__Introduction:__  
We are attemtping to determine the best neighborhood in Pittsburgh for a rat to live.  Rats have all of the same basic needs as humans- an ideal living space to call home, access to food, and safety when out and about.

__Metric__:
In Allegheny County, all dog owners must purchase a dog license.  As dogs are natural predators of rats, it makes sense that the most rat-friendly neighborhood would be the one with the fewest dogs.  Just as humans might factor in crime rate when determining their ideal neighborhood, rats would likely use dog ownership data to gauge how safe a neighborhood is.  

Our dataset will be the "Allegheny County Dog Licenses".  A notable limitation of this data set is that not every dog owner has a dog license, but it's the best data on dog ownership that we can get.  Dog licenses are mandatory and violators will be subject to a fine, but there will always be unlicensed dogs.

This code sorts local zip codes by the number of registered dogs.  

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
data = pd.read_csv('Dogs.csv', delimiter=',')
onlyLocal = data[data['OwnerZip'].astype(str).str.startswith('152')]
zipCount = onlyLocal['OwnerZip'].value_counts()
print(zipCount)

#chart with zip codes on the x axis and number of dog licenses on the y axis

plt.figure(figsize=(30, 8))
zipCount.plot(kind='bar')
plt.title('Number of Dog Licenses in 2024 by Zip Code')
plt.xlabel('Zip Codes')
plt.ylabel('Number of Dog Licenses')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

15237 has the most dog licenses, making it the least rat-safe neighborhood as far.  South Oakland (15275) is the most rat-friendly.

This checks out.  Oakland is a popular neighborhood for students.  Students are less likely to own high-responsibility animals, and landlords in college student-dominated areas are less likely to use pet ownership to entice potential tenants.  Downtown (15219), our overall winner, is also in the top 5.  This is also unsurprising- downtown isn't a very suitable environment for dogs as it lacks green spaces and serious housing.
